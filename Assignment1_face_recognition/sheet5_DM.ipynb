```
def set_data():
  DM = np.load('/content/My_Drive/My Drive/information_system_assignment_1/data_x.npy')
  Y_labels = np.load('/content/My_Drive/My Drive/information_system_assignment_1/data_y.npy')
  
  return DM , Y_labels
```

```
def split_data(split_percent = 0.5):
  DM , Y_labels = set_data()
  DM_train = np.empty((0 , 10304))
  DM_test = np.empty((0 , 10304))
  Y_labels_train = np.empty((0))
  Y_labels_test = np.empty((0))


  if split_percent == 0.5:

    for i in range(400):
      if i % 2 == 0:
        DM_train = np.append(DM_train , DM[i].reshape(1,10304) , axis=0)
        Y_labels_train = np.append(Y_labels_train , Y_labels[i] , axis=0)
      else:
        DM_test = np.append(DM_test , DM[i].reshape(1,10304) , axis=0)
        Y_labels_test = np.append(Y_labels_test , Y_labels[i] , axis=0)
  
    Y_labels_train = Y_labels_train.reshape(200 , 1)
    Y_labels_test = Y_labels_test.reshape(200 , 1)


  else:
    for i in range(400):
      if i % 2 == 0 or (i%10) > 5:
        DM_train = np.append(DM_train , DM[i].reshape(1,10304) , axis=0)
        Y_labels_train = np.append(Y_labels_train , Y_labels[i] , axis=0)
      else:
        DM_test = np.append(DM_test , DM[i].reshape(1,10304) , axis=0)
        Y_labels_test = np.append(Y_labels_test , Y_labels[i] , axis=0)
  
    Y_labels_train = Y_labels_train.reshape(280 , 1)
    Y_labels_test = Y_labels_test.reshape(120 , 1)

  return DM_train , Y_labels_train , DM_test , Y_labels_test
```

```
# centrig data
def center_data(data):
  Mean = (1/data.shape[0]) * (np.sum(data , axis=0 , keepdims=True))
  Z = data - Mean
  return Z
```

## Naive Bayes Classifier

```
def naive_bayes():
  DM_train , Y_labels_train , DM_test , Y_labels_test = split_data()
  naive_bayes = GaussianNB()
  Y_pred = naive_bayes.fit(DM_train , Y_labels_train).predict(DM_test)
  
  print('Y_true labels : ')
  print(Y_labels_test.reshape(1 , 200))

  print('Y_predicted_labels : ')
  print(Y_pred)

  accuracy = accuracy_score(Y_labels_test , Y_pred)
  print('ACCURACY : ' + str(accuracy * 100) + ' %')
  conf_mat = confusion_matrix(Y_labels_test , Y_pred)
  print('CONFUSION MATRIX : ')
  print(conf_mat)
```

##Naive bayes PCA

```
# DM of shape (no of samples , dimensions)
def PCA(DM):
  # shape = (no of examples , 10304)
  Z = center_data(DM)
  # shape = (10304 , 10304)
  cov = (1/DM.shape[0]) * (np.dot(np.transpose(Z) , Z))
  eigvals , eigvecs = np.linalg.eigh(cov)
  eigvals = np.flip(eigvals.reshape(eigvals.shape[0],1) , axis=0)
  eigvecs = np.flip(eigvecs , axis=1)
  
  new_dim = 40
  U_r = eigvecs[:, :new_dim].reshape(eigvecs.shape[0] , new_dim)
  return U_r
```

```
def naive_bayes_PCA():
  DM_train , Y_labels_train , DM_test , Y_labels_test = split_data()
  
  # project train and test data
  eigvecs = PCA(DM_train)
  DM_train_center = center_data(DM_train)
  DM_train_proj = np.dot(DM_train_center , eigvecs)
  DM_test_center = center_data(DM_test)
  DM_test_proj = np.dot(DM_test_center , eigvecs)


  naive_bayes = GaussianNB()
  Y_pred = naive_bayes.fit(DM_train_proj , Y_labels_train).predict(DM_test_proj)
  
  print('Y_true labels : ')
  print(Y_labels_test.reshape(1 , 200))

  print('Y_predicted_labels : ')
  print(Y_pred)

  accuracy = accuracy_score(Y_labels_test , Y_pred)
  print('ACCURACY : ' + str(accuracy * 100) + ' %')
  conf_mat = confusion_matrix(Y_labels_test , Y_pred)
  print('CONFUSION MATRIX : ')
  print(conf_mat)
```

```
naive_bayes_PCA()
```

Accuracy (84.5%) increased due to using PCA for dimensionality reduction and pictures are classified better according to their correct subject for example the first three pictures in subject (1) are predicted correctly to be in subject (1) and confusion matrix values are mainly found at the diagonal which means less error in predicted values.

```
# DM of shape (no of samples , dimensions)
def PCA(DM , no_components):
  # shape = (no of examples , 10304)
  Z = center_data(DM)
  # shape = (10304 , 10304)
  cov = (1/DM.shape[0]) * (np.dot(np.transpose(Z) , Z))
  eigvals , eigvecs = np.linalg.eigh(cov)
  eigvals = np.flip(eigvals.reshape(eigvals.shape[0],1) , axis=0)
  eigvecs = np.flip(eigvecs , axis=1)
  
  new_dim = no_components
  U_r = eigvecs[:, :new_dim].reshape(eigvecs.shape[0] , new_dim)
  return U_r
```

##Decision Tree with PCA

```
def Decision_tree_PCA(no_components):
  DM_train , Y_labels_train , DM_test , Y_labels_test = split_data()
  
  # project train and test data
  eigvecs = PCA(DM_train , no_components)
  DM_train_center = center_data(DM_train)
  DM_train_proj = np.dot(DM_train_center , eigvecs)
  DM_test_center = center_data(DM_test)
  DM_test_proj = np.dot(DM_test_center , eigvecs)

  # information gain is the splitting measure
  Dec_tree = DecisionTreeClassifier(criterion="entropy" , splitter="best")
  Y_pred = Dec_tree.fit(DM_train_proj , Y_labels_train).predict(DM_test_proj)
  
  print('Y_true labels : ')
  print(Y_labels_test.reshape(1 , 200))

  print('Y_predicted_labels : ')
  print(Y_pred)

  accuracy = accuracy_score(Y_labels_test , Y_pred)
  print('ACCURACY : ' + str(accuracy * 100) + ' %')
  conf_mat = confusion_matrix(Y_labels_test , Y_pred)
  print('CONFUSION MATRIX : ')
  print(conf_mat)
```

```
M1 = np.mean(class1 , axis=0)
print('Mean1 : ' + str(M1))
M2 = np.mean(class2 , axis=0)
print('Mean2 : ' + str(M2))
M3 = np.mean(class3 , axis=0)
print('Mean3 : ' + str(M3))
```

```
cov1 = (1/(class1.shape[0])) * np.dot(np.transpose(class1 - M1) , class1 - M1)
print('covariance 1 : ')
print(cov1)
det_cov1 = np.linalg.det(cov1)
print('det of cov1 : ' + str(det_cov1))
print('inv of cov1 : ')
inv_cov1 = np.linalg.inv(cov1)
print(inv_cov1)

cov2 = (1/(class2.shape[0])) * np.dot(np.transpose(class2 - M2) , class2 - M2)
print('covariance 2 : ')
print(cov2)
det_cov2 = np.linalg.det(cov2)
print('det of cov2 : ' + str(det_cov2))
print('inv of cov2 : ')
inv_cov2 = np.linalg.inv(cov2)
print(inv_cov2)

cov3 = (1/(class3.shape[0])) * np.dot(np.transpose(class3 - M3) , class3 - M3)
print('covariance 3 : ')
print(cov3)
det_cov3 = np.linalg.det(cov3)
print('det of cov3 : ' + str(det_cov3))
print('inv of cov3 : ')
inv_cov3 = np.linalg.inv(cov3)
print(inv_cov3)
```
